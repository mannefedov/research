```python
from collections import Counter
from nltk.corpus import stopwords
import re
```

```python
stops = stopwords.words('english')
```

```python
phrases = [x.split('\t') for x in open('phrases.txt').read().splitlines()]
```

```python
phrase_count = Counter()
for p in phrases:
    p = [x for x in p if x not in stops]
    p = [re.sub('(the |a )', '', x) for x in p]
    p = [x for x in p if len(x.split(' ')) > 1]
    phrase_count.update(p)
```

```python
phrase_count.most_common(1000)
```
